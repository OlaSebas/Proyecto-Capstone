## Extract de Ventas

```python

query = """
SELECT 
    lv2.Lv_nro as venta_id,
    lv2.Lv_art as producto_id,
    lv2.Lv_cant as cantidad,
    lv2.Lv_precio as precio_unitario,
    lv2.Lv_linea as total


FROM LVENTA2 AS lv2
LEFT JOIN LVENTA1 AS lv1
    ON lv2.Lv_nro = lv1.Lv_nro
    AND lv2.lv_nro   = lv1.lv_nro

"""

df = pd.read_sql(query, conn)
print("Filas leídas:", len(df))
df.head(20)
```

```python
# --- Convertir a DataFrames ---
df_A = pd.DataFrame(lista_Base_original)
df_B = pd.DataFrame(lista_base_nueva)

# --- Normalizar texto para comparar ---
def normalizar(texto):
    return str(texto).strip().lower()

df_A["norm"] = df_A["descripcion_B"].apply(normalizar)
df_B["norm"] = df_B["descripcion_B"].apply(normalizar)

# --- Comparar por descripción ---
df_merge = pd.merge(df_A, df_B[["id_asignado", "norm"]], on="norm", how="left", suffixes=("_original", "_nueva"))

# --- Resultado final: solo descripción original + id nueva ---
df_resultado = df_merge[["descripcion_B", "id_asignado_nueva","id_asignado_original"]].rename(columns={"id_asignado_nueva": "id_final","id_asignado_original": "id_original"})
```

```python
# --- Corrección manual sobre df_resultado ---
# Asegúrate de que df_resultado ya existe

correcciones = {
    "MENU 6 (POLLO ENTERO + PAPAS + BEBIDA 1.5)": 20,
    "MENU 7 (POLLO ENTERO + PAPAS + BEBIDA 1.5)": 21,
    "AGUA MINERAL CHICA": 25,
    "1/2 POLLO + PAPAS MEDIANAS + BEBIDA 1.5": 22
}

# Aplicar correcciones
df_resultado["id_final"] = df_resultado.apply(
    lambda row: correcciones.get(row["descripcion_B"], row["id_final"]), axis=1
)

# Ver resultado actualizado
df_resultado
```

```python
# Lista de nombres de promociones (normalizadas en minúsculas)
promos = [
    "colacion costillar",
    "colacion para llevar",
    "menu 4 (1/2 pollo + papas medianas",
    "menu 5 (1 pollo entero + papa)",
    "menu 6 (pollo entero+ papas +bebida1.5)",
    "menu 7 (pollo entero+papas+bebida 1.5)",
    "1/2 pollo+papas medianas+bebida 1,5",
    "promo pollo + papa extra grande"
]

# Normalizamos nombres
df_resultado["descripcion_norm"] = df_resultado["descripcion_B"].str.lower().str.strip()

# Marcamos si es promoción
df_resultado["es_promocion"] = df_resultado["descripcion_norm"].isin(promos)

# Creamos columnas separadas
df_resultado["producto_id_final"] = df_resultado.apply(
    lambda x: x["id_final"] if not x["es_promocion"] else None, axis=1
)
df_resultado["promo_id_final"] = df_resultado.apply(
    lambda x: x["id_final"] if x["es_promocion"] else None, axis=1
)

# Eliminamos columna auxiliar si no la necesitas
df_resultado.drop(columns=["descripcion_norm"], inplace=True)

# Mostrar resultado
df_resultado[["descripcion_B", "id_original", "id_final", "es_promocion", "producto_id_final", "promo_id_final"]]
```

```python


# --- Crear diccionario base desde df_resultado ---
dicc_reemplazo_producto = dict(zip(df_resultado["id_original"], df_resultado["producto_id_final"]))
dicc_reemplazo_promo = dict(zip(df_resultado["id_original"], df_resultado["promo_id_final"]))

# --- Asignar columnas nuevas en df ---
df["producto_id_nuevo"] = df["producto_id"].map(dict(zip(df_resultado["id_original"], df_resultado["id_final"])))
df["producto_id_final"] = df["producto_id"].map(dicc_reemplazo_producto)
df["promo_id_final"] = df["producto_id"].map(dicc_reemplazo_promo)

# --- Resultado ---
df[["producto_id", "producto_id_nuevo", "producto_id_final", "promo_id_final"]]
```

## Inicialización de la base de datos sqlite 3 insersión de datos

```python
conn_sqlite = sqlite3.connect(db_path)
cursor = conn_sqlite.cursor()
```

```python
rows = []
for _, row in df.iterrows():
    rows.append((
        int(row["venta_id"]) if pd.notnull(row["venta_id"]) else None,           
        int(row["precio_unitario"]) if pd.notnull(row["precio_unitario"]) else None,    
        int(row["total"]) if pd.notnull(row["total"]) else None,              
        float(row["cantidad"]) if pd.notnull(row["cantidad"]) else None,       
        int(row["promo_id_final"]) if pd.notnull(row["promo_id_final"]) else None,     
        int(row["producto_id_final"]) if pd.notnull(row["producto_id_final"]) else None
    ))
```

```python
insert_sql = """
INSERT INTO venta_detalleVenta
(venta_id, precio_unitario, total, cantidad, promo_id_final, producto_id_final)
VALUES (?, ?, ?, ?, ?, ?)
"""
```

```python
try:
    cursor.executemany(insert_sql, rows)
    conn_sqlite.commit()
    conn_sqlite.close()
    print("Carga a inventario_producto completada.")
except Exception as e:
    print(f"Error durante la carga a inventario_producto: {e}")
```
